### Tiempos de ejecución / uso de memoria

```julia
function suma(N)
    s = 0
    for i ∈ 1:N
        s = s + rand()/i
    end
    return s
end

println(suma(10000000))

@time suma(10000000)
```

```julia
N = 10^6
s = 0
@time begin
    for i ∈ 1:N
        s = s + rand()/i
    end
end
```

```julia
@timev rand(10^6);
```

```julia
using BenchmarkTools
```

```julia
@benchmark rand(10^6)
```
